```python
def missed_text(pokemon):
    print("{} missed!".format(pokemon['Name']))
    sleep(2)
```

```python
def update_status(pokemon1, pokemon2):
    print("\n" + 35*"-")
    print( "| Lvl " + str(pokemon2['Level']) + " " + str(pokemon2['Name']) + " [" + str(pokemon2['Type']) + "]" + (35 - len(str(pokemon2['Level'])) - len(str(pokemon2['Name'])) - len(str(pokemon2['Type'])) - 11)*" " + "|")
    number2 = int(pokemon2['Health']/pokemon2['Max Health'] * 31)
    health2 = number2 * "█"
    if pokemon2['Health'] != 0:
        if health2 == "":
            health2 = 1 * "█"
            number2 -= 1
    print( "| " + health2 + (31 - number2) * "-" + " |")
    print(35 * "-")
    print("\n\n\n\n\n")
    print(10*"\t" + 35*"-")
    print(10*"\t" + "| Lvl " + str(pokemon1['Level']) + " " + str(pokemon1['Name']) + " [" + str(pokemon1['Type']) + "]" + (35 - len(str(pokemon1['Level'])) - len(str(pokemon1['Name'])) - len(str(pokemon1['Type'])) - 11)*" " + "|")
    number1 = int(pokemon1['Health']/pokemon1['Max Health'] * 31)
    health1 = number1 * "█"
    if pokemon1['Health'] != 0:
        if health1 == "":
            health1 = 1 * "█"
            number1 -= 1
    print(10*"\t" + "| " + health1 + (31 - number1) * "-" + " |")
    print(10*"\t" + 35 * "-" + "\n")
```

```python
def accuracy_check(move):
    random_number = round(random(),1)
    if random_number > move['Accuracy']:
        return False
    else:
        return True
```

```python
def attack(pokemon1, pokemon2, attack_move):
    types_table = {('Fire', 'Fire'): 0.5,
                    ('Fire', 'Water'): 0.5,
                    ('Fire', 'Grass'): 2,
                    ('Fire', 'Normal'): 1,
                    ('Water', 'Fire'): 2,
                    ('Water', 'Water'): 0.5,
                    ('Water', 'Grass'): 0.5,
                    ('Water', 'Normal'): 1,
                    ('Grass', 'Fire'): 0.5,
                    ('Grass', 'Water'): 2,
                    ('Grass', 'Grass'): 0.5,
                    ('Grass', 'Normal'): 1,
                    ('Normal', 'Fire'): 1,
                    ('Normal', 'Water'): 1,
                    ('Normal', 'Grass'): 1,
                    ('Normal', 'Normal'): 1}
    check = accuracy_check(attack_move)
    if check == True:
        multiplier = types_table[(attack_move['Type'], pokemon2['Type'])]
        damage_done = (max((pokemon1['Attack'] + (pokemon1['Attack Multiplier']*0.25*pokemon1['Attack']) + (multiplier*attack_move['Base Damage']) - pokemon2['Defence'] - pokemon2['Defence Multiplier']),1))
        pokemon2['Health'] = max(0, pokemon2['Health'] - damage_done)
        if multiplier == 2:
            print("It was super effective!")
            sleep(2)
        elif multiplier == 1:
            pass
        else:
            print("It was not very effective.")
            sleep(2)
    elif check == False:
        missed_text(pokemon1)
    else:
        assert("accuracy_check failure")
    return pokemon2
```

```python
def attack_text(pokemon1, pokemon2, attack_move):
        print("{} has used {} on {}!".format(pokemon1["Name"], attack_move["Name"], pokemon2["Name"]))
        sleep(2)
```

```python
def buff(pokemon, move):
    stage = move['Buff Stage']
    pokemon['{} Multiplier'.format(move['Affected Stat'])] += stage
    pokemon['{} Multiplier'.format(move['Affected Stat'])] = min(pokemon['{} Multiplier'.format(move['Affected Stat'])], 3)
```

```python
def buff_text(pokemon, move):
    print("{} used {}".format(pokemon['Name'], move['Name']))
    sleep(2)
    stage = move['Buff Stage']
    if stage == 1:
        word = "rose"
    elif stage == 2:
        word = "rose sharply"
    elif stage == 3:
        word = "rose very very significantly"
    else:
        assert("Error, stage of buff broken!")
    print("{}'s {} {}.".format(pokemon['Name'], move['Affected Stat'], word))
    sleep(2)
```

```python
def buff_too_much(pokemon, move):
    print("{} used {}".format(pokemon['Name'], move['Name']))
    sleep(2)
    print("But {}'s {} won't go any higher!".format(pokemon['Name'], move['Affected Stat']))
    sleep(2)
```

```python
def debuff(pokemon, move):
    stage = move['Buff Stage']
    pokemon['{} Multiplier'.format(move['Affected Stat'])] -= stage
    pokemon['{} Multiplier'.format(move['Affected Stat'])] = max(pokemon['{} Multiplier'.format(move['Affected Stat'])], -3)
```

```python
def debuff_text(pokemon1, pokemon2, move):
    print("{} has used {} on {}!".format(pokemon1['Name'], move['Name'], pokemon2['Name']))
    sleep(2)
    stage = move['Buff Stage']
    if stage == 1:
        word = "dropped"
    elif stage == 2:
        word = "dropped sharply"
    elif stage == 3:
        word = "dropped very very significantly"
    else:
        assert("Error, stage of debuff broken!")
    print("{}'s {} {}.".format(pokemon2['Name'], move['Affected Stat'], word))
    sleep(2)
```

```python
def debuff_too_much(pokemon1, pokemon2, move):
    print("{} has used {} on {}!".format(pokemon1['Name'], move['Name'], pokemon2['Name']))
    sleep(2)
    print("But {}'s {} won't go any lower!".format(pokemon2['Name'], move['Affected Stat']))
    sleep(2)
```

```python
def user_move(pokemon):
    choice = input("What move do you want your Pokemon to use? \n 1. {} \n 2. {} \n 3. {} \n 4. {} \n Input either 1, 2, 3 or 4.\n".format((pokemon['Movesets']["1"]['Name'] + " [" + pokemon['Movesets']["1"]['Type'] + "]"), (pokemon['Movesets']["2"]['Name'] + " [" + pokemon['Movesets']["2"]['Type'] + "]"), (pokemon['Movesets']["3"]['Name'] + " [" + pokemon['Movesets']["3"]['Type'] + "]"), (pokemon['Movesets']["4"]['Name'] + " [" + pokemon['Movesets']["4"]['Type'] + "]")))
    
    while choice != "1" and choice != "2" and choice != "3" and choice != "4":
        choice = input("Please try again! \n What move do you want your Pokemon to use? \n 1. {} \n 2.{} \n 3. {} \n 4. {} \n Input either 1, 2, 3 or 4.".format((pokemon['Movesets']["1"]['Name'] + " [" + pokemon['Movesets']["1"]['Type'] + "]"), (pokemon['Movesets']["2"]['Name'] + " [" + pokemon['Movesets']["2"]['Type'] + "]"), (pokemon['Movesets']["3"]['Name'] + " [" + pokemon['Movesets']["3"]['Type'] + "]"), (pokemon['Movesets']["4"]['Name'] + " [" + pokemon['Movesets']["4"]['Type'] + "]")))
        
    return pokemon['Movesets'][choice]
```

```python
def check_remaining_pokemon(trainer): #takes a dictionary "trainer" and returns an integer for the number of remaining pokemon
    remaining_pokemon = 0
    for pokemon in trainer['Pokemon']:
        if pokemon['Health'] == 0:
            continue
        else:
            remaining_pokemon += 1
    return remaining_pokemon
```

```python
def change_pokemon(trainer, pokemon): #takes a dictionary "trainer" and a "pokemon" and sets the global variable for pokemon chosen by "trainer"
    for each_pokemon in range(len(trainer['Pokemon'])):
        if trainer['Pokemon'][each_pokemon] == pokemon:
            if trainer['Name'] == "Saiyan":
                global Saiyan_pokemon_choice 
                Saiyan_pokemon_choice = each_pokemon
                print("I choose you, {}".format(trainer['Pokemon'][each_pokemon]['Name']))
                sleep(2)
            else:
                global Enemy_pokemon_choice 
                Enemy_pokemon_choice = each_pokemon
                print("{} sent out {}".format(trainer['Name'],trainer['Pokemon'][each_pokemon]['Name']))
                sleep(2)
        else:
            continue
```

```python
def fainting(trainer): #takes a dictionary "trainer" and decides what to do after trainer's pokemon faints
    if check_remaining_pokemon(trainer) == 1:
        for pokemon in trainer['Pokemon']:
            if pokemon['Health'] == 0:
                continue
            else:
                change_pokemon(trainer, pokemon)
    elif check_remaining_pokemon(trainer) == 0:
        global battle_over 
        battle_over = True
    else:
        available_pokemon = {}
        available_pokemon_types = {}
        for pokemon_index in range(len(trainer['Pokemon'])):
            if trainer['Pokemon'][pokemon_index]['Health'] == 0:
                continue
            else:
                available_pokemon[str(pokemon_index+1)] = trainer['Pokemon'][pokemon_index]['Name']
                available_pokemon_types[str(pokemon_index+1)] = trainer['Pokemon'][pokemon_index]['Type']
        
        for pokemon_index,types_index in zip(available_pokemon,available_pokemon_types):
            print("{} : {} [{}]".format(pokemon_index, available_pokemon[pokemon_index], available_pokemon_types[types_index]))
        
        choice = input("Please input your next choice of pokemon: ")
        while choice not in (available_pokemon.keys()):
            choice = input("Please input a number. Try Again: ")
        choice =int(choice)-1
        change_pokemon(trainer, trainer['Pokemon'][choice])
```

```python
def reset(trainer):
    for pokemon in trainer['Pokemon']:
        pokemon['Health'] = pokemon['Max Health']
        pokemon['Attack Multiplier'] = 0
        pokemon['Defence Multiplier'] = 0
    return trainer

def reset_pokemon_multiplier(pokemon):
    pokemon['Attack Multiplier'] = 0
    pokemon['Defence Multiplier'] = 0
    return pokemon
    
def reset_multiplier(trainer):
    for pokemon in trainer['Pokemon']:
        pokemon = reset_pokemon_multiplier(pokemon)
    return trainer
```

```python
def battle(trainer, Saiyan):
    global Saiyan_pokemon_choice
    Saiyan_pokemon_dict = {}
    Saiyan_pokemon_types = {}
    for pokemon_index in range(len(Saiyan['Pokemon'])):
        if Saiyan['Pokemon'][pokemon_index]['Health'] != 0:
            Saiyan_pokemon_dict[pokemon_index+1] = Saiyan['Pokemon'][pokemon_index]['Name']
            Saiyan_pokemon_types[pokemon_index+1] = Saiyan['Pokemon'][pokemon_index]['Type']
            
    for pokemon_index,types_index in zip(Saiyan_pokemon_dict,Saiyan_pokemon_types):
        print("{} : {} [{}]".format(pokemon_index, Saiyan_pokemon_dict[pokemon_index], Saiyan_pokemon_types[types_index]))
    
    Saiyan_pokemon_choice = int(input("Choose your starting pokemon: ")) -1
    
    start_of_battle_text(trainer['Name'], trainer['Pokemon'][0], Saiyan['Name'], Saiyan['Pokemon'][Saiyan_pokemon_choice])
    global battle_over 
    battle_over = False
    own_turn = True
    global Enemy_pokemon_choice 
    Enemy_pokemon_choice= 0
    while battle_over == False:
        # fight
        while trainer['Pokemon'][Enemy_pokemon_choice]['Health'] > 0 and Saiyan['Pokemon'][Saiyan_pokemon_choice]['Health'] > 0:
            if own_turn:
                #Saiyan Turn
                update_status(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice])
                turn_choice = input("1. Fight\n2. Change Pokemon\n")
                if turn_choice == "1":
                    move = user_move(Saiyan['Pokemon'][Saiyan_pokemon_choice])
                    if move['Move Type'] == "Attack":
                        attack_text(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice], move)
                        trainer['Pokemon'][Enemy_pokemon_choice] = attack(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice], move)
                        update_status(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice])
                    elif move['Move Type'] == "Buff":
                        if Saiyan['Pokemon'][Saiyan_pokemon_choice]["{} Multiplier".format(move['Affected Stat'])] == 3:
                            buff_too_much(Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                        else:
                            buff(Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                            buff_text(Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                    elif move['Move Type'] == "Debuff":
                        if trainer['Pokemon'][Enemy_pokemon_choice]["{} Multiplier".format(move['Affected Stat'])] == -3:
                            debuff_too_much(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice], move)
                        else:
                            debuff(trainer['Pokemon'][Enemy_pokemon_choice], move)
                            debuff_text(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice], move)
                    else:
                        assert("Move Type error!")
                elif turn_choice == "2":
                    Saiyan['Pokemon'][Saiyan_pokemon_choice] = reset_pokemon_multiplier(Saiyan['Pokemon'][Saiyan_pokemon_choice])
                    other_pokemon = {}
                    other_pokemon_types = {}
                    for pokemon_index in range(len(Saiyan['Pokemon'])):
                        if Saiyan['Pokemon'][pokemon_index]['Health'] == 0 or Saiyan['Pokemon'][pokemon_index]['Name'] == Saiyan['Pokemon'][Saiyan_pokemon_choice]['Name']:
                            continue
                        else:
                            other_pokemon[str(pokemon_index+1)] = Saiyan['Pokemon'][pokemon_index]['Name']
                            other_pokemon_types[str(pokemon_index+1)] = Saiyan['Pokemon'][pokemon_index]['Type']
                    print("\n")
                    
                    if len(other_pokemon.keys()) == 0:
                        turn_choice = input("You have no more pokemons left! Please choose again.")
                    else:
                        for pokemon_index,types_index in zip(other_pokemon,other_pokemon_types):
                            print("{} : {} [{}]".format(pokemon_index, other_pokemon[pokemon_index], other_pokemon_types[types_index]))

                        choice = input("Please input your next choice of pokemon: ")
                        while choice not in (other_pokemon.keys()):
                            choice = input("Please input a number. Try Again: ")
                        choice = int(choice)-1
                        change_pokemon(Saiyan, Saiyan['Pokemon'][choice])
                else:
                    turn_choice = input("Incorrect input. Please input 1 to fight or 2 to change pokemon\n\n1.Fight\n2.Change Pokemon\n")
            else:
                #Troll Miltank
                if trainer['Pokemon'][Enemy_pokemon_choice]['Name'] == 'Miltank':
                    if trainer['Pokemon'][Enemy_pokemon_choice]['Health'] > 120:
                        if trainer['Pokemon'][Enemy_pokemon_choice]['Attack Multiplier'] == 3:
                            if trainer['Pokemon'][Enemy_pokemon_choice]['Defence Multiplier'] == 3:
                                move = trainer['Pokemon'][Enemy_pokemon_choice]['Movesets']["4"]
                            else:
                                move = trainer['Pokemon'][Enemy_pokemon_choice]['Movesets']["{}".format(2 * randint(1,2))]
                        elif trainer['Pokemon'][Enemy_pokemon_choice]['Defence Multiplier'] == 3:
                            move = trainer['Pokemon'][Enemy_pokemon_choice]['Movesets']["{}".format(randint(3,4))]
                        else:
                            move = trainer['Pokemon'][Enemy_pokemon_choice]['Movesets']["{}".format(randint(2,4))]
                    else:
                        move = trainer['Pokemon'][Enemy_pokemon_choice]['Movesets']['1']
                else:
                    move = ai_choose_move(trainer['Pokemon'][Enemy_pokemon_choice])
                #Opponent Turn
                if move['Name'] == 'Milk Drink':
                    print("Miltank used Milk Drink!")
                    sleep(2)
                    trainer['Pokemon'][Enemy_pokemon_choice]['Health'] = 300
                    print("Miltank is now at full health")
                    sleep(2)
                    
                elif move['Move Type'] == "Attack":
                    attack_text(trainer['Pokemon'][Enemy_pokemon_choice], Saiyan['Pokemon'][Saiyan_pokemon_choice], move)            
                    Saiyan['Pokemon'][Saiyan_pokemon_choice] = attack(trainer['Pokemon'][Enemy_pokemon_choice], Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                elif move['Move Type'] == "Buff":
                    if trainer['Pokemon'][Enemy_pokemon_choice]["{} Multiplier".format(move['Affected Stat'])] == 3:
                        buff_too_much(trainer['Pokemon'][Enemy_pokemon_choice], move)
                    else:
                        buff(trainer['Pokemon'][Enemy_pokemon_choice], move)
                        buff_text(trainer['Pokemon'][Enemy_pokemon_choice], move)
                elif move['Move Type'] == "Debuff":
                    if Saiyan['Pokemon'][Saiyan_pokemon_choice]["{} Multiplier".format(move['Affected Stat'])] == -3:
                        debuff_too_much(trainer['Pokemon'][Enemy_pokemon_choice], Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                    else:
                        debuff(Saiyan['Pokemon'][Enemy_pokemon_choice], move)
                        debuff_text(trainer['Pokemon'][Enemy_pokemon_choice], Saiyan['Pokemon'][Saiyan_pokemon_choice], move)
                else:
                    assert("Move Type Error")
            own_turn = not own_turn
            if trainer['Pokemon'][Enemy_pokemon_choice]['Health'] == 0:
                print("{} has fainted.".format(trainer['Pokemon'][Enemy_pokemon_choice]['Name']))
                sleep(2)
                fainting(trainer)
                own_turn = True
                if battle_over == True:
                    win_text(trainer)
                    return True
            elif Saiyan['Pokemon'][Saiyan_pokemon_choice]['Health'] == 0:
                update_status(Saiyan['Pokemon'][Saiyan_pokemon_choice], trainer['Pokemon'][Enemy_pokemon_choice])
                sleep(2)
                print("{} has fainted.".format(Saiyan['Pokemon'][Saiyan_pokemon_choice]['Name']))
                sleep(2)
                fainting(Saiyan)
                own_turn = True
                if battle_over == True:
                    lose_text()
                    return False
            else:
                continue
    
```
